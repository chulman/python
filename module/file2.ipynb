```python
import pickle
import os, shutil
import zipfile
color = ['red', 'green']

f = open('color.pickle','wb')
pickle.dump(color, f)
f.close()
```

```python
# file, dest
shutil.copy('./color.pickle', '../')
# 기존 파일이 있으면 move 안됨
shutil.move('./color.pickle', '../')
```

```python
# 디렉토리 순회
# os.walk(path)

for folder_name, sub_folders, files in os.walk('../'):
    print("folder_names:" + folder_name)
    
    for sub_folder in sub_folders:
        print("sub_folder:" + sub_folder)
    
    for file in files:
        print("file :" + file)
```

```python
# zip 파일 압축 및 풀기
examZip = zipfile.ZipFile('exam.zip', 'w')
examZip.write('color.pickle', compress_type=zipfile.ZIP_DEFLATED)
examZip.close()
```

```python
examZip2 = zipfile.ZipFile('exam.zip')
examZip2.extractall()
examZip2.close()
```
